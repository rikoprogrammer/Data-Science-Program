```python
# Replacing missing data with the column mean
train['Age'].fillna(value=train['Age'].mean(), inplace=True)
test['Age'].fillna(value=test['Age'].mean(), inplace=True)
```

```python
# Selecting features for the model
x_train = train.iloc[:,[2,4,5,6,7]].values
x_test = test.iloc[:,[1,3,4,5,6]].values
y_train = train['Survived'].values
y_test = submission['Survived'].values
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
#Encodind the independent variable sex
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x_train = ct.fit_transform(x_train)
print(x_train)
```

```python
ct2 = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x_test = ct2.fit_transform(x_test)
print(x_test)
```

```python
#Modeling, i use logist regression to predict survived variable

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(x_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(x_test)

print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))

```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
